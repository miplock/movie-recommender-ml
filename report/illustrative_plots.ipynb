```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
rmse_data = pd.read_csv("parametrized_rmse_evaluation_results/all_algs_rmse_data.csv") 
components = rmse_data['n_components']
```

```python
plt.plot(components, rmse_data['NMF_z'], label = 'NMF, Zero', color = 'lime')
plt.plot(components, rmse_data['NMF_mm'], label = 'NMF, Movie mean', color = 'lightgreen')
plt.plot(components, rmse_data['NMF_um'], label = 'NMF, User mean', color = 'darkgreen') 
plt.title('RMSE for NMF model') 
plt.ylabel('RMSE')
plt.xlabel('Number of components')
plt.xticks(components)   
plt.legend() 
plt.show()
```

```python
plt.plot(components, rmse_data['SVD1_z'], label = 'SVD1, Zero', color = 'blue')
plt.plot(components, rmse_data['SVD1_mm'], label = 'SVD1, Movie mean', color = 'cornflowerblue')
plt.plot(components, rmse_data['SVD1_um'], label = 'SVD1, User mean', color = 'midnightblue')  
plt.title('RMSE for SVD1 model') 
plt.ylabel('RMSE')
plt.xlabel('Number of components')
plt.xticks(components)  
plt.legend() 
plt.show()
```

```python
plt.plot(components, rmse_data['SVD2_z'], label = 'SVD2, Zero', color = 'red')
plt.plot(components, rmse_data['SVD2_mm'], label = 'SVD2, Movie mean', color = 'lightcoral')
plt.plot(components, rmse_data['SVD2_um'], label = 'SVD2, User mean', color = 'maroon') 
plt.title('RMSE for SVD2 model') 
plt.ylabel('RMSE')
plt.xlabel('Number of components')
plt.xticks(components)   
plt.legend() 
plt.show()
```

```python
plt.plot(components, rmse_data['SGD'], label = 'SGD', color = 'orange')    
plt.title('RMSE for SGD model') 
plt.ylabel('RMSE')
plt.xticks(components)
plt.xlabel('Matrix rank')
plt.legend() 
plt.show()
```

```python
plt.plot(components, rmse_data['WMF'], label = 'WMF', color = 'magenta') 
plt.title('RMSE for WMF model')   
plt.ylabel('RMSE')
plt.xticks(components)
plt.xlabel('Matrix rank')
plt.legend() 
plt.show()
```

```python
#plt.plot(components, rmse_data['NMF_mm'], label = 'NMF, Movie mean', color = 'lightgreen')
plt.plot(components, rmse_data['NMF_um'], label = 'NMF, User mean', color = 'darkgreen')  
#plt.plot(components, rmse_data['SVD1_mm'], label = 'SVD1, Movie mean', color = 'cornflowerblue')
plt.plot(components, rmse_data['SVD1_um'], label = 'SVD1, User mean', color = 'midnightblue') 
#plt.plot(components, rmse_data['SVD2_mm'], label = 'SVD2, Movie mean', color = 'lightcoral')
plt.plot(components, rmse_data['SVD2_um'], label = 'SVD2, User mean', color = 'maroon') 
plt.plot(components, rmse_data['SGD'], label = 'SGD', color = 'orange') 
plt.plot(components, rmse_data['WMF'], label = 'WMF', color = 'magenta')    
plt.legend() 
plt.title('RMSE for the best case of each model')
plt.xlabel('n\n (matrix rank for SGD & WMF,\n number of components for NMF, SVD1 & SVD 2)')
plt.ylabel('RMSE')
plt.xticks(components)
plt.show()
```
